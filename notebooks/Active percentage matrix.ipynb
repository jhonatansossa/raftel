```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
import pyspark.sql.functions as F
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df_spark = catalog.load("aras_a@spark")
```

```python
df_spark.orderBy(F.col("DAY").asc(), F.col("SEC").asc()).show()
```

```python
APM = df_spark.select(
    'Ph1',
    'Ph2',
    'Ir1',
    'Fo1',
    'Fo2',
    'Di3',
    'Di4',
    'Ph3',
    'Ph4',
    'Ph5',
    'Ph6',
    'Co1',
    'Co2',
    'Co3',
    'So1',
    'So2',
    'Di1',
    'Di2',
    'Te1',
    'Fo3',
    'DAY',
    'SEC',
    F.expr("stack(2, LR1, LR2) as LR")
).orderBy(F.col("DAY").asc(), F.col("SEC").asc())
```

```python
APM.show()
```

```python
APM = APM.groupby(F.col("LR").alias("id")).agg(
    (F.avg("Ph1")*100).alias("AP_Ph1"),
    (F.avg("Ph2")*100).alias("AP_Ph2"),
    (F.avg("Ir1")*100).alias("AP_Ir1"),
    (F.avg("Fo1")*100).alias("AP_Fo1"),
    (F.avg("Fo2")*100).alias("AP_Fo2"),
    (F.avg("Di3")*100).alias("AP_Di3"),
    (F.avg("Di4")*100).alias("AP_Di4"),
    (F.avg("Ph3")*100).alias("AP_Ph3"),
    (F.avg("Ph4")*100).alias("AP_Ph4"),
    (F.avg("Ph5")*100).alias("AP_Ph5"),
    (F.avg("Ph6")*100).alias("AP_Ph6"),
    (F.avg("Co1")*100).alias("AP_Co1"),
    (F.avg("Co2")*100).alias("AP_Co2"),
    (F.avg("Co3")*100).alias("AP_Co3"),
    (F.avg("So1")*100).alias("AP_So1"),
    (F.avg("So2")*100).alias("AP_So2"),
    (F.avg("Di1")*100).alias("AP_Di1"),
    (F.avg("Di2")*100).alias("AP_Di2"),
    (F.avg("Te1")*100).alias("AP_Te1"),
    (F.avg("Fo3")*100).alias("AP_Fo3")
).orderBy("LR")
```

```python
data = [
    (1, "Other"),
    (2, "Going Out"),
    (3, "Preparing Breakfast"),
    (4, "Having Breakfast"),
    (5, "Preparing Lunch"),
    (6, "Having Lunch"),
    (7, "Preparing Dinner"),
    (8, "Having Dinner"),
    (9, "Washing Dishes"),
    (10, "Having Snack"),
    (11, "Sleeping"),
    (12, "Watching TV"),
    (13, "Studying"),
    (14, "Having Shower"),
    (15, "Toileting"),
    (16, "Napping"),
    (17, "Using Internet"),
    (18, "Reading Book"),
    (19, "Laundry"),
    (20, "Shaving"),
    (21, "Brushing Teeth"),
    (22, "Talking on the Phone"),
    (23, "Listening to Music"),
    (24, "Cleaning"),
    (25, "Having Conversation"),
    (26, "Having Guest"),
    (27, "Changing Clothes")
]
```

```python
columns = ["id", "Activities"]
activities = spark.createDataFrame(data, columns)
```

```python
APM = APM.join(activities, "id", "INNER").drop("id")
```

```python
APM.toPandas()
```

```python
APM_pandas = APM_pandas.drop([20])
```
