```python
labels = [
    'Brushing_Teeth',
    'Changing_Clothes',
    'Cleaning',
    'Going_Out',
    'Having_Breakfast',
    'Having_Conversation',
    'Having_Dinner',
    'Having_Guest',
    'Having_Lunch',
    'Having_Shower',
    'Having_Snack',
    'Laundry',
    'Listening_to_Music',
    'Napping',
    'Other',
    'Preparing_Breakfast',
    'Preparing_Dinner',
    'Preparing_Lunch',
    'Reading_Book',
    'Shaving',
    'Sleeping',
    'Studying',
    'Talking_on_the_Phone',
    'Toileting',
    'Using_Internet',
    'Washing_Dishes',
    'Watching_TV'
]


val_counts = y.sum(axis=0)
plt.figure(figsize=(15,10))
ax = sns.barplot(val_counts, alpha=0.8)
plt.title("Labels per Classes")
plt.xlabel("Various Label Type")
plt.ylabel("Counts of the Labels")
rects = ax.patches
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height+5, label, ha="center", va="bottom")
plt.show()
```
