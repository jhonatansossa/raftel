#### No Augmented Visualization

```python
labels = [
    'Brushing_Teeth',
    'Changing_Clothes',
    'Cleaning',
    'Going_Out',
    'Having_Breakfast',
    'Having_Conversation',
    'Having_Dinner',
    'Having_Guest',
    'Having_Lunch',
    'Having_Shower',
    'Having_Snack',
    'Laundry',
    'Listening_to_Music',
    'Napping',
    'Other',
    'Preparing_Breakfast',
    'Preparing_Dinner',
    'Preparing_Lunch',
    'Reading_Book',
    'Shaving',
    'Sleeping',
    'Studying',
    'Talking_on_the_Phone',
    'Toileting',
    'Using_Internet',
    'Washing_Dishes',
    'Watching_TV'
]

val_counts = non_augmented_y.sum(axis=0)
plt.figure(figsize=(15,10))
ax = sns.barplot(val_counts, alpha=0.8)
plt.title("Instances per Activities")
plt.xlabel("Activities")
plt.ylabel("Instances")
plt.ylim(0, 17500)
rects = ax.patches
for rect, label, value in zip(rects, labels, val_counts):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height+100, f"{label} \n ({value})", ha="center", va="bottom", rotation=90, fontsize=10)
plt.show()
```

#### Augmented Visualization

```python
val_counts = augmented_y_train.sum(axis=0)
plt.figure(figsize=(15,10))
ax = sns.barplot(val_counts, alpha=0.8)
plt.title("Instances per Activities")
plt.xlabel("Activities")
plt.ylabel("Instances")
plt.ylim(0, 17500)
rects = ax.patches
for rect, label, value in zip(rects, labels, val_counts):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height+100, f"{label} \n ({value})", ha="center", va="bottom", rotation=90, fontsize=10)
plt.show()
```
